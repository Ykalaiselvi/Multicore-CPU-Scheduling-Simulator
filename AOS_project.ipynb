```python
import gradio as gr
import matplotlib.pyplot as plt

# ---------------- FCFS Scheduling ----------------
def fcfs(processes, burst):
    n = len(processes)
    waiting = [0]*n
    turnaround = [0]*n

    for i in range(1,n):
        waiting[i] = waiting[i-1] + burst[i-1]

    for i in range(n):
        turnaround[i] = waiting[i] + burst[i]

    avg_wait = sum(waiting)/n
    avg_turn = sum(turnaround)/n

    # Gantt Chart
    time = 0
    fig, ax = plt.subplots()
    for i in range(n):
        ax.barh(1, burst[i], left=time)
        ax.text(time + burst[i]/2, 1, processes[i], ha='center', color='white')
        time += burst[i]

    ax.set_title("FCFS Gantt Chart")
    ax.set_yticks([])
    ax.set_xlabel("Time")
    plt.close()

    result = f"Waiting Time: {waiting}\nTurnaround Time: {turnaround}\nAverage Waiting Time: {avg_wait:.2f}\nAverage Turnaround Time: {avg_turn:.2f}"
    return result, fig

# -------------- INPUT HANDLER ----------------
def run_scheduler(pnames, bursts):
    processes = pnames.split(',')
    burst = list(map(int, bursts.split(',')))
    return fcfs(processes, burst)

# ---------------- UI ----------------
ui = gr.Interface(
    fn=run_scheduler,
    inputs=[
        gr.Textbox(label="Process Names (Example: P1,P2,P3,P4)"),
        gr.Textbox(label="Burst Times (Example: 5,3,8,6)")
    ],
    outputs=[
        gr.Textbox(label="Result"),
        gr.Plot(label="Gantt Chart")
    ],
    title="CPU Scheduling Simulator (FCFS)",
    description="Advanced Operating System Scheduler Simulator running in Google Colab"
)

ui.launch()
```
